```python
# Jaipur Super Market Billing System

# Product dictionary (Key   -> Product Number , Value -> Tuple (Product Name, Product Price))

products = {
    1: ("Rice", 60),
    2: ("Milk", 50),
    3: ("Sugar", 45),
    4: ("Oil", 150),
    5: ("Soap", 30),
    6: ("Tea", 100),
    7: ("Coffee", 150),
    8: ("Toothpaste", 100),
    9: ("Chocolate", 120),
    10:("Cold-Drink", 55)
}

cart = {}  # Cart dictionary to store selected items
total = 0  # Variables for total bill
discount=0 # Variables for discount

#Customer Interaction 
print("Welcome to Japiur Super Market","\n","="*40)
name=input("Please enter you name:").capitalize()
number=input("Please enter your contact number:")
print("\n\nHello ",name," \nbelow is the list of available products:-\n\n")



for p in products: # Displaying available products
    print(p, products[p][0], "‚Çπ", products[p][1])
print("\n","="*80)
    
while True: # Allows user to select multiple products
    choice = int(input ( " \nPlease select the product number to into your cart:"))
    if choice not in products:  # Validate product number
        print("invalid choice of product")
        continue
    
    qnt = int(input(f"Please select the quantity for {products[choice][0]}: ") )
    if choice in cart:
        cart[choice]+=qnt
    else:
        cart[choice]=qnt
    more=input("Do you want to add more items into the cart (y/n)?")
    if more.lower() == "n":
        break

input("\nPress any key to generate invoice\n")
# generating invoice
print("="*80,"\n Jaipur Super Market Invoice\n","="*80)
print ("Name :",name,"\nContact number : ",number)
from datetime import datetime

today = datetime.now()
print("Date : ",today.strftime("%A, %d %B %Y"))
print("\n","S.N |  " ,"Product |","Quantity |","Price |","Total ")
q=1
for i in cart:
    item_total = cart[i]*products[i][1]
    total+=item_total 
    print(" ",q," |", products[i][0],"     ",cart[i],"      ",products[i][1],"    ",item_total)
    q+=1


#Discount Calculations
if total>=5000:
    discount=0.10
elif total>=2000:
    discount=0.05


#final payable amount calculation 
print("\n","-"*40,"\n Total Quantity: ",sum(cart.values()),"\n Amount: ",total,"\n Discount: ",int(discount*100),"%","\n Total Payable: ‚Çπ", total -(total*discount)," only")
print("-"*40,"\nüôè Thank you for shopping with us! üôè\n Jaipur Super Market")    
```
